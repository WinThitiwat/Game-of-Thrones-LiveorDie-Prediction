# Run predictive analysis

```python


X = GOT.loc[:,['book1_A_Game_Of_Thrones', 'book2_A_Clash_Of_Kings',
       'book3_A_Storm_Of_Swords', 'book4_A_Feast_For_Crows',
       'book5_A_Dance_with_Dragons', 'isMarried', 'isNoble', 'age',
       'numDeadRelations', 'popularity',
       'great_house', 'm_house',
       'great_title', 'm_title']].values
y = GOT['isAlive'].values

names = ['book1_A_Game_Of_Thrones', 'book2_A_Clash_Of_Kings',
       'book3_A_Storm_Of_Swords', 'book4_A_Feast_For_Crows',
       'book5_A_Dance_with_Dragons', 'isMarried', 'isNoble', 'age',
       'numDeadRelations', 'popularity',
       'great_house', 'm_house',
       'great_title', 'm_title']

lasso = Lasso(0)
lasso_coef = lasso.fit(X,y).coef_
plt.plot(range(len(names)), lasso_coef)
plt.xticks(range(len(names)), names, rotation=60)
plt.ylabel("Coef")
```

```python
from sklearn.preprocessing import StandardScaler
GOT_data = GOT.loc[:, ['book1_A_Game_Of_Thrones', 
       'book4_A_Feast_For_Crows',
         'great_house',
       'numDeadRelations', 'popularity'
       
       ]]
temp_col = GOT_data.columns
# 'm_house','m_title' 'great_title', 'm_title' 'book2_A_Clash_Of_Kings', 'age', great_house'
GOT_target = GOT.loc[:, "isAlive"]

scaler = StandardScaler()
scaler.fit(GOT_data)
X_scaled = scaler.transform(GOT_data)
GOT_data = pd.DataFrame(X_scaled)
GOT_data.columns = temp_col
X_train, X_test, y_train, y_test = train_test_split(GOT_data, 
                                                    GOT_target, 
                                                    test_size=0.1,
                                                   random_state=508,
                                                   stratify=GOT_target)
```

```python

got_smc = smc.Logit(GOT_target, GOT_data)
results = got_smc.fit()

results.summary()
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

print(knn.score(X_train, y_train))
print(knn.score(X_test, y_test))
```

# Hyperparameter

```python
from sklearn.model_selection import GridSearchCV

grid_params = {
    'n_neighbors': np.arange(1,30)
}

knn_gs = GridSearchCV(KNeighborsClassifier(),
                     grid_params,
                      return_train_score =False,
                      scoring=None,
                      cv=3
                    )

# Fitting the model based on the training data
knn_gs_results = knn_gs.fit(X_train, y_train)

print(knn_gs_results.best_score_)
print(knn_gs_results.best_estimator_)
print(knn_gs_results.best_params_)
```

```python
bp = knn_gs_results.best_params_
knn = KNeighborsClassifier( n_neighbors= bp["n_neighbors"])

# cv_knn_10 = cross_val_score(knn, GOT_data, GOT_target,scoring='accuracy' ,cv=kf)
# cv_knn_10
knn.fit(X_train, y_train)

y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)

print('Training Score',knn.score(X_train, y_train))
print('Testing Score:', knn.score(X_test, y_test))

print('Training AUC Score', roc_auc_score(y_train, y_train_pred).round(4))
print('Testing AUC Score', roc_auc_score(y_test, y_test_pred).round(4))
```

# Cross-validating our Model

```python
# Cross-Validating the model with three folds
cv_knn_optimal_3 = cross_val_score(knn,
                                      GOT_data,
                                      GOT_target,
                                      cv = 3)


print(cv_knn_optimal_3)


print('\nAverage: ',
      pd.np.mean(cv_knn_optimal_3).round(3),
      '\nMinimum: ',
      min(cv_knn_optimal_3).round(3),
      '\nMaximum: ',
      max(cv_knn_optimal_3).round(3))
```

```python
print(confusion_matrix(y_true = y_test,
                       y_pred = y_test_pred))


# Visualizing the confusion matrix
labels = ['Not Alive', 'Alive']

cm = confusion_matrix(y_true = y_test,
                      y_pred = y_test_pred)


sns.heatmap(cm,
            annot = True,
            xticklabels = labels,
            yticklabels = labels,
            cmap = 'Blues')


plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion matrix of the classifier')
plt.show()
```
